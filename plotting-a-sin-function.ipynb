```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace (0,4.*np.pi,65530)
```

```python
y = np.sin (x)
```

```python
plt.plot (x,y);
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('This is the x-axis')
axes.set_ylabel('This is the y-axis')
axes.set_title('Sine Function Plot');
plt.show ()
```
